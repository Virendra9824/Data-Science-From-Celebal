```python
import pandas as pd
import numpy as np
```

```python
data = {
    'col1': ['789xyz', '123', 'hello', '!@!', ' ', '987', '123abc', '   ', 'world!'],
}
df = pd.DataFrame(data)
df.head(9)
```

### T3.1: Check whether alphabetic values present in a given column of a DataFrame.

```python
def check_alpha(column):
    return column.apply(lambda word: any(ch.isalpha() for ch in word))


df['has_alpha'] = check_alpha(df['col1'])
df
```

### T3.2: Check whether only numeric values are present in a given column

```python
def check_numeric(column):
    return column.apply(lambda x: x.isdigit())


df['is_numeric'] = check_numeric(df['col1'])
df
```

### T3.3: Check whether only space is present in a given column

```python
def check_space(column):
    return column.apply(lambda x: x.isspace())


df['is_space'] = check_space(df['col1'])
df
```

### T3.4: Extract only non-alphanumeric characters from the specified column

```python
def extract_non_alphanumeric(column):
    return column.apply(lambda word: ''.join([ch for ch in word if not ch.isalnum()]))


df['non_alphanumeric'] = extract_non_alphanumeric(df['col1'])
df
```

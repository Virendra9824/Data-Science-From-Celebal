```python
df1=df.head(20)
```

# Bar Graph between Age and Marital-status:

```python
plt.bar(df1['marital-status'],df1['age'] , color = "magenta", edgecolor="y")
plt.xlabel("Marital-status", color ="r", fontsize=18)
plt.ylabel("Age", color ="r", fontsize=18)
plt.xticks(rotation = 15)
plt.title("Marital-Status vs Age", color="m", fontsize=20)
plt.show()
```

# Scatter Plot

```python
plt.scatter(df['relationship'],df['capital-gain'] , color = "blue", edgecolor="y")
plt.xlabel("Marital-status", color ="r", fontsize=18)
plt.ylabel("Age", color ="r", fontsize=18)
plt.xticks(rotation = 15)
plt.title("relationship vs capital-gain", color="m", fontsize=20)
plt.show()
```

# Line Plot Between Hours-per-week and Age

```python
plt.plot(df1['hours-per-week'],df1['age'], marker="*") 
plt.xlabel("hours-per-week", color ="r", fontsize=18) 
plt.ylabel("Age", color ="r", fontsize=18) 
plt.xticks(rotation = 15) 
plt.title("Hours-per-week vs Age", color="m", fontsize=20) 
plt.show()
```

### Pie plot

```python
ex = [0.001]*df1['age']
plt.pie(df1['age'], labels=df1['relationship'], explode=ex,radius=1.8, autopct='%.0f%%')
# plt.legend()
plt.show()
```

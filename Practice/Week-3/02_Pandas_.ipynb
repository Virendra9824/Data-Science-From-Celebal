```python
python_df
```

```python
python_df.rename(columns={'count': 'NumRespondants', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondants'])*100
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False)
```
